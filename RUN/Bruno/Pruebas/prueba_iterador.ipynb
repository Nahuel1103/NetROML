Este notebook es para probar como funciona un iterador externo cuando se lo llama desde una clase. La idea de esto es que varias instancias de una clase puedan ir tomando datos coordinadamente de un iterador sin repetirse.

Esta prueba la hago pensando en dos cosas:
1) Ver como pasarle una H por paso a nuestro env.
2) Ver si funciona para paralelizar a futuro.

```python
class ToyEnv:
    def __init__(self, H_iterator=None, name="env"):
        self.H_iterator = H_iterator
        self.name = name
        self.H = None

    def _get_H(self):
        if self.H_iterator is None:
            raise ValueError("No hay iterador definido.")
        try:
            self.H = next(self.H_iterator)
        except StopIteration:
            self.H = None
            print(f"{self.name}: iterador agotado")
        return self.H
    

    def step(self):
        H = self._get_H()
        print(f"{self.name} obtiene H = {H}")
        return H
    

    
```

```python
# -----------------------
# Probemos el comportamiento
# -----------------------
H_list = [10, 20, 30, 40, 50]
H_iter = iter(H_list)  # iterador compartido

# Creamos dos instancias que comparten el iterador
env1 = ToyEnv(H_iterator=H_iter, name="env1")
env2 = ToyEnv(H_iterator=H_iter, name="env2")

# Llamadas intercaladas
env1.step()  # debería sacar 10
env2.step()  # debería sacar 20
env1.step()  # debería sacar 30
env2.step()  # debería sacar 40
env1.step()  # debería sacar 50
env2.step()  # iterador agotado
```

**Conclusión:**

Parece que funciona como esperaría, por lo que esto se podría utilizar en mi env real. Sería como tener una pila externa a mi env y que se vayan tomando elementos (POP) de ella.
